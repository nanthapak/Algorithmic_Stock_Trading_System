```python
import yfinance as yf
data_train = yf.download('GOOG','2014-01-01','2017-12-31')
data_test=yf.download('GOOG','2018-01-01','2018-12-31')
data=yf.download('GOOG','2014-01-01','2018-12-31')
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(data_train['Adj Close'],color='green',label='training period')
plt.plot(data_test['Adj Close'],color='red',label='testing period')
plt.title('GOOG')
plt.legend()
plt.legend(loc='best',shadow=True,fontsize='small') #หาที่ว่างเองโดยอัตโนมัติ
plt.grid()
```
